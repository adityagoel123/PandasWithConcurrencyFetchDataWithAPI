```python
## Modular style || Function to hit to 3rd party API and return the required result.


def fetchMeActualAadhaarNumber(thisRow):
    #print(thisRow)
    #print("############1")
    thisReferenceNumber = thisRow["COD_AADHAAR_NO"]
    #print(thisReferenceNumber)
    #thisReferenceNumber = 999672072017
    payload = '{\n\t\"referenceKey\":\"'+ str(thisReferenceNumber) + '\",\n\t\"apiKey\":\"paymentBank\",\n\t\"requestId\":\"uniqueskey\"\n}'
    response = requests.request("POST", url, headers=headers, data = payload, verify=False)
    responseAPIInBytes = response.text.encode('utf8')
    fix_bytes_value = responseAPIInBytes.replace(b"'", b'"')
    my_json = json.load(io.BytesIO(fix_bytes_value))  
    #print(my_json)
    #print("############2")
    checkIfAadhaarGotGeneratedSuccesfully = 'data' in my_json
    if checkIfAadhaarGotGeneratedSuccesfully:
        #print('c key is present')
        result = my_json['data']['uid']
    else:
        result = my_json['meta']['description']    
    #print(result)
    #print("############3")
    return result
    #dfWithDataFetched = dfWithDataFetched.append({'COD_ACCT_NO' : thisRow['COD_ACCT_NO'],
    #                                             'COD_CUST_ID' : thisRow['COD_CUST_ID'],
    #                                             'FLG_CUST_TYP' : thisRow['FLG_CUST_TYP'],
    #                                             'COD_AADHAAR_NO' : thisRow['COD_AADHAAR_NO'],
    #                                             'ACTUAL_AADHAR_NO' : result,
    #                                             }, ignore_index = True)
    #break    
```
